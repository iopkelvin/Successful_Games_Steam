```python
import pandas as pd
pd.set_option('display.max_columns', 100)
import numpy as np
```

## GOOGLE APPS

```python
google_play = pd.read_csv('google-play-store-apps/googleplaystore.csv')
print("['Countries dropped', 'Android Ver', 'Current Ver', 'Genres', 'Price']")
google_play.drop(['Android Ver', 'Current Ver', 'Genres', 'Price'], axis=1, inplace=True)
print("Google Play dataset: ", google_play.shape)
google_play.sample(3)
```

## APP STORE

```python
app_store = pd.read_csv('appstore_games.csv')
print("Columns dropped: ['Subtitle' ,'ID', 'URL', 'Icon URL', 'Description', 'Developer', 'Languages', 'Genres']")
app_store.drop(['Subtitle' ,'ID', 'URL', 'Icon URL', 'Description', 'Developer', 'Languages', 'Genres'], axis=1, inplace=True)
print("App Store dataset 2016: ", app_store.shape)
app_store.sample(3)
```
